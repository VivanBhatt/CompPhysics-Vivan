```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**7*np.sin(x)

def firstderivative_approx(f, x, h):
    return (f(x-2*h) - 8*f(x-h) + 8*f(x+h) - f(x+2*h)) / (12*h)

x = np.pi/4
true_firstderivative = x**7*np.cos(x)+7*x**6*np.sin(x)

h_values = np.logspace(-16, 0, 100)
errors = [abs(firstderivative_approx(f, x, h) - true_firstderivative) for h in h_values]

plt.loglog(h_values, errors)
plt.xlabel('Step size (h)')
plt.ylabel('Absolute Error')
plt.title('Error vs Step Size')
plt.grid(True)
plt.show()
```

```python
epsilon_machine = np.finfo(np.float64).eps
print("Machine epsilon for double precision:", epsilon_machine)
```

```python
errors_expected = [h**4+epsilon_machine/h for h in h_values]
plt.loglog(h_values, errors, label='Approximated Error')
plt.loglog(h_values, errors_expected, label='Expected Error')
plt.title('Expected and actual errors')
plt.xlabel('Step size (h)')
plt.grid(True)
plt.legend()
plt.show()
```
